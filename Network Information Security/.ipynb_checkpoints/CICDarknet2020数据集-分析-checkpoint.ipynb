# Libraries importing and configuration 库导入和配置

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from matplotlib import font_manager
my_font = font_manager.FontProperties(fname='../Fonts/SourceHanSerifSC-Medium.otf', size=20)
```

```python
dataset_folder = "./outputs/" 
img_folder = ""

plt.rcParams['figure.figsize'] = (18,6)   # 图片像素
plt.rcParams["savefig.format"] = 'png'
plt.rcParams["savefig.dpi"] = 600   # 分辨率 
plt.rcParams.update({'font.size': 18})
```

```python
# 导入经过预处理的CSV文件
samples = pd.read_csv(dataset_folder + 'darknet_dataset_processed.csv')
```

```python
samples.head()
```

# Traffic origin 原始流量

下图显示了良性和暗网流量样本数量的详细信息

```python
df = samples.groupby("Label").count()
fig, ax = plt.subplots(figsize=(8,8))

df.plot(y="Label.1", ax=ax, shadow=False, kind="pie", fontsize=18,autopct=lambda x: int(df["Label.1"].values.sum()*x/100))
plt.ylabel("")
plt.tight_layout()
plt.savefig('distr_dados_ref')
plt.show()
```

## Labels probability distribution 标签概率分布

```python
sns.histplot(data=samples, x="Label", stat="frequency")
plt.xlabel('')
plt.show()
```

## Access hour 访问小时数

```python
# Non-Tor NonVPN Tor VPN
plot_graph_Non_Tor = samples[samples['Label'] == "Non-Tor"]
plot_graph_NonVPN = samples[samples['Label'] == "NonVPN"]
plot_graph_Tor = samples[samples['Label'] == "Tor"]
plot_graph_VPN = samples[samples['Label'] == "VPN"]
sns.kdeplot(plot_graph_Non_Tor['hour'], shade=True)
sns.kdeplot(plot_graph_NonVPN['hour'], shade=True)
sns.kdeplot(plot_graph_Tor['hour'], shade=True)
sns.kdeplot(plot_graph_VPN['hour'], shade=True)
plt.xlabel("小时",font_properties=my_font)
plt.ylabel("访问密度",font_properties=my_font)
plt.legend(['Non-Tor', 'NonVPN', 'Tor', 'VPN'])
plt.savefig(img_folder + 'hour_relationship2')
plt.show()
```

```python
samples['Label'] == ("Tor")
```

```python
plot_graph_darknet = samples[(samples['Label'] == "Tor") | (samples['Label'] =="VPN")]
plot_graph_benign = samples[(samples['Label'] == "Non-Tor") | (samples['Label'] == "NonVPN")]
sns.kdeplot(plot_graph_darknet['hour'], shade=True)
sns.kdeplot(plot_graph_benign['hour'], shade=True)
plt.xlabel("小时",font_properties=my_font)
plt.ylabel("访问密度",font_properties=my_font)
plt.legend(['Darknet', 'Benigna'])
plt.savefig(img_folder + 'hour_relationship2')
plt.show()
```

# Traffic characterization

## Type of traffic on Darknet

```python
sns.histplot(data=samples[(samples['Label'] == "Tor") | (samples['Label'] =="VPN")], x="Label.1", stat="probability")
plt.xlabel('')
plt.ylabel("比例",font_properties=my_font)
plt.show()
```

## Type of traffic outside the Darknet

```python
sns.histplot(data=samples[(samples['Label'] == "Non-Tor") | (samples['Label'] == "NonVPN")], x="Label.1", stat="probability")
plt.ylabel("比例",font_properties=my_font)
plt.show()
```
